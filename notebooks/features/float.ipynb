# Float

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.impute import MissingIndicator

%matplotlib inline

from smp.features.features import Loader

loader = Loader()
```

## Avg Daily Profile Visit Duration in seconds

```python
variable = "Avg Daily Profile Visit Duration in seconds"

print(loader.train[variable].head(20))
```

```python
X = np.log(1 + loader.train[variable].values)
y = np.log(1 + loader.train["Num of Profile Likes"])
```

```python
_ = plt.hist(X, bins=50, alpha=0.5, density=True)
_ = plt.title(variable)
_ = plt.xlabel(variable)
_ = plt.ylabel("Frequency")
```

```python
_ = plt.scatter(X, y)

_ = plt.title(variable)
_ = plt.xlabel(variable)
_ = plt.ylabel("Num of Profile Likes")
```
