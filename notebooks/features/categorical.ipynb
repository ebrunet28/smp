# Categorical

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.impute import MissingIndicator

%matplotlib inline

from smp.features.features import Loader

loader = Loader()
```

## Profile Cover Image Status

```python
variable = "Profile Cover Image Status"

print(loader.train[variable].head(20))
```

```python
X = loader.train[variable]
y = np.log(1 + loader.train["Num of Profile Likes"])

_ = plt.hist(y[X == 'Set'], bins=100, alpha=0.5, density=True, label="Set")
_ = plt.hist(y[X == 'Not set'], bins=100, alpha=0.5, density=True, label="Not set")

_ = plt.title(variable)
_ = plt.xlabel("Num of Profile Likes")
_ = plt.ylabel("Frequency")
_ = plt.legend()
```

## User Language

```python
variable = "User Language"

print(loader.train[variable].head(20))
```

```python
X = loader.train[variable]
y = np.log(1 + loader.train["Num of Profile Likes"])

_ = plt.hist(y[X == 'en'], bins=100, alpha=0.5, density=True, label="English")
_ = plt.hist(y[X == 'es'], bins=100, alpha=0.5, density=True, label="Spanish")
_ = plt.hist(y[~X.isin(['en', 'es'])], bins=100, alpha=0.5, density=True, label="Other")

_ = plt.title(variable)
_ = plt.xlabel("Num of Profile Likes")
_ = plt.ylabel("Frequency")
_ = plt.legend()
```
