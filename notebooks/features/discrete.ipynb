# Discrete

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.impute import MissingIndicator

%matplotlib inline

from smp.features.features import Loader

loader = Loader()
```

## UTC Offset

```python
variable = "UTC Offset"

print(loader.train[variable].head(20))
```

```python
X = loader.train[variable].values
y = np.log(1 + loader.train["Num of Profile Likes"])
```

```python
_ = plt.hist(X, bins=50, alpha=0.5, density=True)
_ = plt.title(variable)
_ = plt.xlabel(variable)
_ = plt.ylabel("Frequency")
```

```python
_ = plt.hist(y[X < 0], bins=100, alpha=0.5, density=True, label="< 0")
_ = plt.hist(y[X >= 0], bins=100, alpha=0.5, density=True, label=">= 0")

_ = plt.title(variable)
_ = plt.xlabel("Num of Profile Likes")
_ = plt.ylabel("Frequency")
_ = plt.legend()
```

```python
_ = plt.scatter(X, y)

_ = plt.title(variable)
_ = plt.xlabel(variable)
_ = plt.ylabel("Num of Profile Likes")

## Num of Followers
```
