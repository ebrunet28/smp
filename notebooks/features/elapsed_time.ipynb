# Elapsed Time

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

from smp.features.features import Loader

loader = Loader()
```

## Profile Creation Timestamp

```python
variable = "Profile Creation Timestamp"

print(loader.train[variable].head(20))
```

```python
X = loader.train[variable]
X = pd.to_datetime(X).dt.date
most_recent = X.max()
X = X.apply(lambda x: (most_recent - x).days)
y = np.log(1 + loader.train["Num of Profile Likes"])
```

```python
_ = plt.hist(X, bins=50, alpha=0.5, density=True)
_ = plt.title(variable)
_ = plt.xlabel(variable)
_ = plt.ylabel("Frequency")
```

```python
_ = plt.scatter(X, y)

_ = plt.title(variable)
_ = plt.xlabel(variable)
_ = plt.ylabel("Num of Profile Likes")
```
