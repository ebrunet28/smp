# RBG

```python
import matplotlib.pyplot as plt
import numpy as np

from smp.features.features import Loader

loader = Loader()
```

## Profile Text Color

```python
variable = "Profile Text Color"

print(loader.train[variable].head(20))
```

```python
X = loader.train[variable]
y = np.log(1 + loader.train["Num of Profile Likes"])
```

```python
_ = plt.hist(y[X == '0084b4'], bins=100, alpha=0.5, density=True, label="0084b4")
_ = plt.hist(y[X != '0084b4'], bins=100, alpha=0.5, density=True, label="!= 0084b4")

_ = plt.title(variable)
_ = plt.xlabel("Num of Profile Likes")
_ = plt.ylabel("Frequency")
_ = plt.legend()
```

```python
_ = plt.hist(y[X.isin(['0084b4', '1da1f2'])], bins=100, alpha=0.5, density=True, label="Top 2")
_ = plt.hist(y[~X.isin(['0084b4', '1da1f2'])], bins=100, alpha=0.5, density=True, label="Bottom")

_ = plt.title(variable)
_ = plt.xlabel("Num of Profile Likes")
_ = plt.ylabel("Frequency")
_ = plt.legend()
```

```python
_ = plt.hist(y[X.isin(['0084b4', '1da1f2', '009999'])], bins=100, alpha=0.5, density=True, label="Top 3")
_ = plt.hist(y[~X.isin(['0084b4', '1da1f2', '009999'])], bins=100, alpha=0.5, density=True, label="Bottom")

_ = plt.title(variable)
_ = plt.xlabel("Num of Profile Likes")
_ = plt.ylabel("Frequency")
_ = plt.legend()
```
